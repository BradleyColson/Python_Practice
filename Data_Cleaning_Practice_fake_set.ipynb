```python
df = pd.read_csv(r"C:\Users\Bradl\downloads\data_to_be_cleaned.csv")
df
```

```python
#Displays how many totals vs how many nulls
#Result shows 2 null client, 2 null amount, 3 null ClientTier
df.info()
```

```python
#Make a copy of df so we have the original raw file
cleaned_df = df.copy()
```

```python
 #Drop client nulls
cleaned_df.dropna(subset=["Client"], inplace=True)
```

```python
#Did the client nulls get droped
cleaned_df
```

```python
#Drop duplicates with understanding that transaction ID isn't the same
cleaned_df.drop_duplicates(subset=["Client", "Beneficiary", "AmountUSD", "TransactionDateTime", "ClientTier"], inplace=True)
```

```python
cleaned_df
```

```python
#Assuming it this is 10000 rows long, what characters do we not want
#Remove special characters
#Fixed all lines except line 5
cleaned_df["Client"].str.replace("[a-zA-Z0-9]", "", regex=True).unique()
```

```python
cleaned_df["Client"].str.strip("<>.*- ")
```

```python
cleaned_df["Client"] = cleaned_df["Client"].str.strip("<>.*- ")
```

```python
cleaned_df
```

```python
#Fix Harry*Potter
cleaned_df["Client"] = cleaned_df["Client"].str.replace("*", " ")
cleaned_df
```

```python
#Fix any ALL CAPS
cleaned_df["Client"] = cleaned_df["Client"].str.upper()
cleaned_df["Beneficiary"] = cleaned_df["Beneficiary"].str.upper()
cleaned_df
```

```python
#AmountUSD has some nan, can we infer what the values might have been? Add mean,median.

ax = cleaned_df['AmountUSD'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
ax.set_ylabel('Frequency')
ax.set_xlabel('AmountUSD')
ax.legend()
plt.show()
```

```python

# AmountUSD has some nan, can we infer what the values might have been? Add mean,median.

mean_value = cleaned_df["AmountUSD"].mean()
median_value = cleaned_df["AmountUSD"].median()

ax = cleaned_df['AmountUSD'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
ax.set_ylabel('Frequency')
ax.set_xlabel('AmountUSD')

ax.axvline(mean_value, color='darkblue', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
ax.axvline(median_value, color='darkred', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')

ax.legend()
plt.show()
```
